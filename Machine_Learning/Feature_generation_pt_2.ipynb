```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python
tdf = pd.read_csv('tdf.csv',index_col=0)
idfb = pd.read_csv('idfb.csv',index_col=0)
bdfb = pd.read_csv('bdfb.csv',index_col=0)
tdfb = pd.read_csv('tdfb.csv',index_col=0)
hdfb = pd.read_csv('hdfb.csv',index_col=0)
ddfb = pd.read_csv('ddfb.csv',index_col=0)
df = pd.read_csv('dict_df.csv',index_col=0)
```

## 1 b inhib

```python
idfb.head()
```

```python
idfb1 = pd.melt(idfb, id_vars=['match_id2'])
idfb1.variable = idfb1.variable.astype(int)
```

```python
idfb1.loc[(idfb1['value'] == 'Inhib'),]
```

```python
idfb1.drop(137,inplace=True)
idfb1.loc[(idfb1['value'] == 'Inhib'),]
```

```python
idfb1.sort_values(['match_id2', 'variable'],inplace=True)
idfb1['value'] = idfb1['value'].astype(float).round()
idfb1.reset_index(drop=True,inplace=True)
    
idfb1 = idfb1[(idfb1 != 0).all(1)]
idfb1.sort_values(['match_id2','value'],inplace=True)
idfb1['count'] = idfb1.groupby('match_id2').cumcount() + 1
```

```python
idfb1.rename(columns={'count':'b_inhib_count','value':'minute'},inplace=True)
idfb1.drop('variable',inplace=True,axis=1)
```

```python
print(idfb1.isnull().any())
idfb1.reset_index(drop=True,inplace=True)
idfb1.tail()
```
